```python
points = torch.tensor([[4,0],[5,3],[2,1]]).to(torch.float)
#though point is a 2D tensor,every [] inside the tensor is alos a tensor , in storage it is actually 1D [4,0,5,3,2,1]
                                                                                                       #[0,1,2,3,4,5]
#offset is the staring point in the memory
#size/shape is the number of elements
#stride is the number of elements to skip to go in next ror/column from current offset 
second_point = points[1]
#accessing an element(i,j) in 2D tensor means : storage offset+ stride[0]*i + stride[1]*j , offset is usually zero for a 2D tensor
print(second_point.size()) #number of elements
print(second_point.storage_offset()) #offset is the staring point , in this case second point[5,3] starts from 2nd index
print(points.stride()) # in this case going to new row , we have to add 2 with offset and 1 if we want to go to next column
print(second_point.stride())
```
